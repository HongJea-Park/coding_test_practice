````python
from timeit import timeit


def search_list(array: list, x):

    return [idx for (idx, element) in enumerate(array) if element == x]


v = [17, 92, 18, 33, 58, 7, 33, 42]
print(search_list(v, 18))  # [2]
print(search_list(v, 33))  # [3, 6]
print(search_list(v, 900))  # []

print(timeit(
    stmt='search_list(v, 33)',
    setup='from __main__ import search_list, v',
    number=1))
````

<br>
```
stu_no = [39, 14, 67, 105]
stu_name = ["Justin", "John", "Mike", "Summer"]
```


````python
from timeit import timeit


stu_no = [39, 14, 67, 105]
stu_name = ["Justin", "John", "Mike", "Summer"]


def find_stu(stu_no: list, stu_name: list, number: int):

    for no, name in zip(stu_no, stu_name):
        if no == number:
            return name
    return "?"


print(find_stu(stu_no, stu_name, 67))
print(find_stu(stu_no, stu_name, 19))
print(timeit(
    stmt='find_stu(stu_no, stu_name, 67)',
    setup='from __main__ import find_stu, stu_no, stu_name',
    number=1
))
````

### Problem 8.1

<br>일반적인 선택 정렬 알고리즘을 사용해서 리스트 `[2, 4, 5, 1, 3]` 을 정렬하는 과정을 적어 보세요.

````python
from timeit import timeit


def sel_sort_ascending_order(array: list):

    for i in range(len(array)-1):
        min_idx = i
        for j in range(i+1, len(array)):
            if array[j] < array[min_idx]:
                min_idx = j
        array[i], array[min_idx] = array[min_idx], array[i]
    return array


a = [2, 4, 5, 1, 3]
print(sel_sort_ascending_order(a))
print(timeit(
    stmt='sel_sort_ascending_order(a)',
    setup='from __main__ import sel_sort_ascending_order, a',
    number=1
))
````

## Problem 9

<br>일반적인 삽입 정렬 알고리즘을 사용해서 리스트 `[2, 4, 5, 1, 3]` 을 정렬하는 과정을 적어 보세요.

````python
from timeit import timeit


def ins_sort_ascending_order(array: list):

    for i in range(1, len(array)):
        element = array[i]
        j = i-1
        while j >= 0 and array[j] > element:
            array[j+1] = array[j]
            j -= 1
        array[j+1] = element
    return array


a = [2, 4, 5, 1, 3]
print(ins_sort_ascending_order(a))
print(timeit(
    stmt='ins_sort_ascending_order(a)',
    setup='from __main__ import ins_sort_ascending_order, a',
    number=1
))
````

```python
# Program 10-2
# 병합 정렬
# 입력: 리스트 a
# 출력: 없음(입력으로 주어진 a가 정렬됨)

def merge_sort(a):
    n = len(a)
    # 종료 조건: 정렬할 리스트의 자료 개수가 한 개 이하이면 정렬할 필요가 없음
    if n <= 1:
        return
    # 그룹을 나누어 각각 병합 정렬을 호출하는 과정
    mid = n // 2  # 중간을 기준으로 두 그룹으로 나눔
    g1 = a[:mid]
    g2 = a[mid:]
    merge_sort(g1)  # 재귀 호출로 첫 번째 그룹을 정렬
    merge_sort(g2)  # 재귀 호출로 두 번째 그룹을 정렬
    # 두 그룹을 하나로 병합
    i1 = 0
    i2 = 0
    ia = 0
    while i1 < len(g1) and i2 < len(g2):
        if g1[i1] < g2[i2]:
            a[ia] = g1[i1]
            i1 += 1
            ia += 1
        else:
            a[ia] = g2[i2]
            i2 += 1
            ia += 1
    # 아직 남아 있는 자료들을 결과에 추가
    while i1 < len(g1):
        a[ia] = g1[i1]
        i1 += 1
        ia += 1
    while i2 < len(g2):
        a[ia] = g2[i2]
        i2 += 1
        ia += 1

d = [6, 8, 3, 9, 10, 1, 2, 4, 7, 5]
merge_sort(d)
print(d)
```

<br>프로그램 10-2에서 다룬 정렬 알고리즘은 숫자를 작은 수에서 큰 수 순서로 나열하는 오름차순 정렬이었습니다. 
<br>오른차순 정렬을 큰 수에서 작은 수 순서로 나열하는 내림차순 정렬로 바꾸려면 프로그램의 어느 부분을 바꿔야 할까요? 

````python
import numpy as np
from timeit import timeit


def merge_sort_descending(array: list):

    if len(array) <= 1:
        return

    mid = len(array) // 2
    g1 = array[:mid].copy()
    g2 = array[mid:].copy()
    g_idx1 = 0
    g_idx2 = 0
    array_idx = 0

    merge_sort_descending(g1)
    merge_sort_descending(g2)

    def insert(array_idx: int, g: list, g_idx: int):
        array[array_idx] = g[g_idx]
        g_idx += 1
        array_idx += 1
        return array_idx, g_idx

    while g_idx1 < len(g1) and g_idx2 < len(g2):
        if g1[g_idx1] > g2[g_idx2]:
            array_idx, g_idx1 = insert(array_idx, g1, g_idx1)
        else:
            array_idx, g_idx2 = insert(array_idx, g2, g_idx2)
    while g_idx1 < len(g1):
        array_idx, g_idx1 = insert(array_idx, g1, g_idx1)
    while g_idx2 < len(g2):
        array_idx, g_idx2 = insert(array_idx, g2, g_idx2)


def merge_sort(a):
    n = len(a)
    if n <= 1:
        return
    mid = n // 2
    g1 = a[:mid]
    g2 = a[mid:]
    merge_sort(g1)
    merge_sort(g2)
    i1 = 0
    i2 = 0
    ia = 0
    while i1 < len(g1) and i2 < len(g2):
        if g1[i1] > g2[i2]:
            a[ia] = g1[i1]
            i1 += 1
            ia += 1
        else:
            a[ia] = g2[i2]
            i2 += 1
            ia += 1
    while i1 < len(g1):
        a[ia] = g1[i1]
        i1 += 1
        ia += 1
    while i2 < len(g2):
        a[ia] = g2[i2]
        i2 += 1
        ia += 1


array = np.arange(10000)
np.random.shuffle(array)

print('time cost of merge sort')
print(timeit(
    stmt='merge_sort_descending(array)',
    setup='from __main__ import merge_sort_descending, array',
    number=1
))
print(array)

print('-'*25)

np.random.shuffle(array)
print('time cost of selection sort')
print(timeit(
    stmt='sel_sort_descending_order(array)',
    setup='from __main__ import sel_sort_descending_order, array',
    number=1
))
print(sel_sort_descending_order(array))

print('-'*25)

np.random.shuffle(array)
print('time cost of insert sort')
print(timeit(
    stmt='ins_sort_descending_order(array)',
    setup='from __main__ import ins_sort_descending_order, array',
    number=1
))
print(ins_sort_descending_order(array))

print('-'*25)

np.random.shuffle(array)
print('time cost of program 10-2')
print(timeit(
    stmt='merge_sort(array)',
    setup='from __main__ import merge_sort, array',
    number=1
))
print(array)
````

### Problem 11.1

<br>지금까지 배운 네 가지 정렬 알고리즘 말고도 훨씬 많은 정렬 알고리즘이 있습니다. 
<br>그 중 하나인 거품 정렬(Bubble sort)을 줄 서기로 비유하면 다음과 같습니다. 
<br>다음 과정을 읽고 리스트 `[2, 4, 5, 1, 3]`이 정렬되는 과정을 알고리즘으로 적어보세요.

````python
import numpy as np
from timeit import timeit


def bubble_sort_ascending(array: list):

    while True:

        change = False

        for idx in range(len(array)-1):

            if array[idx] > array[idx+1]:
                array[idx], array[idx+1] = array[idx+1], array[idx]
                change = True

        if not change:
            break

    return array


def bubble_sort_descending(array: list):

    while True:

        change = False

        for idx in range(len(array)-1):

            if array[idx] < array[idx+1]:
                array[idx], array[idx+1] = array[idx+1], array[idx]
                change = True

        if not change:
            break

    return array


array = np.arange(10000)
np.random.shuffle(array)
print(timeit(
    stmt='bubble_sort_ascending(array)',
    setup='from __main__ import bubble_sort_ascending, array',
    number=1
))
print(array)

print('-'*25)

np.random.shuffle(array)
print(timeit(
    stmt='bubble_sort_descending(array)',
    setup='from __main__ import bubble_sort_descending, array',
    number=1
))
print(array)
````

## Problem 12
### Problem 12.1

<br> 다음 과정을 참고하여 재귀 호출을 사용한 이분 탐색 알고리즘을 만들어 보세요.
<br>

1. 주어진 탐색 대상이 비어 있다면 탐색할 필요가 없습니다. (종료 조건)
2. 찾는 값과 주어진 탐색 대상의 중간 위치 값을 비교합니다.
3. 찾는 값과 중간 위치 값이 같다면 결과값으로 중간 위치 값을 돌려줍니다.
4. 찾는 값이 중간 위치 값보다 크다면 중간 위치의 오른쪽을 대상으로 이분 탐색 함수를 재귀 호출합니다.
5. 찾는 값이 중간 위치 값보다 작다면 중간 위치의 왼쪽을 대상으로 이분 탐색 함수를 재귀 호출합니다.

````python
import numpy as np
from timeit import timeit


def binary_search_recursive(array: list, value: int):

    if not array:
        return -1

    center_idx = len(array) // 2

    if value == array[center_idx]:
        return array[center_idx]
    elif value > array[center_idx]:
        return binary_search_recursive(array[center_idx+1:], value)
    else:
        return binary_search_recursive(array[:center_idx], value)


def binary_search_for(array: list, value: int):

    start_idx, center_idx, end_idx = 0, len(array) // 2, len(array)

    while end_idx - start_idx > 1:
        if value > array[center_idx]:
            start_idx = center_idx
        else:
            end_idx = center_idx
        center_idx = (start_idx+end_idx) // 2

    if value == array[start_idx] or value == array[end_idx]:
        return value
    else:
        return -1


array = sorted(
    np.random.choice(np.arange(1000), size=500, replace=False).tolist())
value = np.random.randint(low=0, high=1000)

print(timeit(
    stmt='binary_search_recursive(array, value)',
    setup='from __main__ import binary_search_recursive, array, value',
    number=1
))
print(f"Find {value} index in array with recursive: "
      f"{binary_search_recursive(array, value)}")

print('-'*25)

print(timeit(
    stmt='binary_search_for(array, value)',
    setup='from __main__ import binary_search_for, array, value',
    number=1
))
print(f"Find {value} index in array with for loop: "
      f"{binary_search_for(array, value)}")
````
