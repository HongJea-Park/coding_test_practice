## 문제

지원자가 지원서에 입력한 4가지의 정보와 획득한 코딩테스트 점수를 하나의 문자열로 구성한 값의 배열 info, 개발팀이 궁금해하는 문의조건이 문자열 형태로 담긴 배열 query가 매개변수로 주어질 때,
각 문의조건에 해당하는 사람들의 숫자를 순서대로 배열에 담아 return 하도록 solution 함수를 완성해 주세요. <br>

## 제한사항

- info 배열의 크기는 1 이상 50,000 이하입니다.
- info 배열 각 원소의 값은 지원자가 지원서에 입력한 4가지 값과 코딩테스트 점수를 합친 개발언어 직군 경력 소울푸드 점수 형식입니다.
    - 개발언어는 cpp, java, python 중 하나입니다.
    - 직군은 backend, frontend 중 하나입니다.
    - 경력은 junior, senior 중 하나입니다.
    - 소울푸드는 chicken, pizza 중 하나입니다.
    - 점수는 코딩테스트 점수를 의미하며, 1 이상 100,000 이하인 자연수입니다.
    - 각 단어는 공백문자(스페이스 바) 하나로 구분되어 있습니다.
- query 배열의 크기는 1 이상 100,000 이하입니다.
- query의 각 문자열은 (조건) X 형식입니다.
    - (조건)은 개발언어 and 직군 and 경력 and 소울푸드 형식의 문자열입니다.
    - 언어는 cpp, java, python, - 중 하나입니다.
    - 직군은 backend, frontend, - 중 하나입니다.
    - 경력은 junior, senior, - 중 하나입니다.

    - '-' 표시는 해당 조건을 고려하지 않겠다는 의미입니다.
    - X는 코딩테스트 점수를 의미하며 조건을 만족하는 사람 중 X점 이상 받은 사람은 모두 몇 명인 지를 의미합니다.
    - 각 단어는 공백문자(스페이스 바) 하나로 구분되어 있습니다.

# 풀이
<br>
query마다 info를 탐색하면 효율성 테스트를 통과할 수 없다.<br>
query는 4개의 요소에 categorical feature로 구성되어 있기 때문에 categorical feature로 group을 만들어 따로 저장해두면 query가 들어올 때 해당 group 내에서만 지원자들의 점수를 확인하면 된다.<br>
이 때 group별로 sort해둔 뒤 binary search를 통해 탐색을 진행하면 query에서 요구하는 점수 이상 지원자의 수를 빠르게 탐색할 수 있다.<br>
<br>
<br>

```python
import bisect
from itertools import product


def key_combination(lang, posi, exp, food):
    keyList = []
    langList = [lang, '-']
    posiList = [posi, '-']
    expList = [exp, '-']
    foodList = [food, '-']
    for lang, posi, exp, food in product(langList, posiList, expList, foodList):
        keyList.append(''.join([lang, posi, exp, food]))
    return keyList


def solution(info, query):
    infoDict, answer = {}, []
    langList = ['cpp', 'java', 'python', '-']
    posiList = ['backend', 'frontend', '-']
    expList = ['junior', 'senior', '-']
    foodList = ['chicken', 'pizza', '-']
    for lang, posi, exp, food in product(langList, posiList, expList, foodList):
        infoDict[''.join([lang, posi, exp, food])] = []
    
    for i in info:
        lang, posi, exp, food, score = i.split()
        keyList = key_combination(lang, posi, exp, food)
        for key in keyList:
            infoDict[key].append(int(score))
    
    for key, value in infoDict.items():
        infoDict[key] = sorted(value, reverse=True)

    for q in query:
        q = q.replace('and ', '')
        lang, posi, exp, food, score = q.split()
        key = ''.join([lang, posi, exp, food])
        if key in infoDict.keys():
            answer.append(bisect.bisect_left(infoDict[key], int(score)))
        else:
            answer.append(0)

    return answer
```
