# Sherlock and Cost

[링크](https://www.hackerrank.com/challenges/sherlock-and-cost/problem)

<br>

## 문제 설명

In this challenge, you will be given an array $B$ and must determine an array $A$. <br>
There is a special rule: 

- For all $i$, $A[i] \le B[i]$. <br>

<br>

That is,  $A[i]$ can be any number you choose such that $1 \le A[i] \le B[i]$. <br>
Your task is to select a series of $A[i]$ given $B[i]$ such that the sum of the absolute difference of consecutive pairs of $A$ is maximized.<br>
This will be the array's cost, and will be represented by the variable $S$ below. <br>

<br>

The equation can be written: <br>

$\displaystyle S = \sum_{i}^N |A[i] - A[i-1] |$ <br>

For example, if the array $B=[1, 2, 3]$, we know that $1 \le A[1] \le 1$, $1 \le A[2] \le 2$, and $1 \le A[3] \le 3$. <br>
Arrays meeting those guidelines are: <br>

```
[1,1,1], [1,1,2], [1,1,3]
[1,2,1], [1,2,2], [1,2,3]
```

<br>

Our calculations for the arrays are as follows:

```
|1-1| + |1-1| = 0	|1-1| + |2-1| = 1	|1-1| + |3-1| = 2
|2-1| + |1-2| = 2	|2-1| + |2-2| = 1	|2-1| + |3-2| = 2
```

<br>

The maximum value obtained is 2.

<br>

### Function Description

Complete the cost function in the editor below. <br>
It should return the maximum value that can be obtained.

cost has the following parameter(s):

- B: an array of integers

<br>

### Input Format

The first line contains the integer $t$, the number of test cases.

Each of the next  pairs of lines is a test case where:
- The first line contains an integer , the length of 
- The next line contains  space-separated integers 

<br>

### Constraints

- $1 \le t \le 20$
- $1 \le n \le 10^5$
- $1 \le B[i] \le 100$

<br>

### Output Format

For each test case, print the maximum sum on a separate line.

<br>

### Sample Input

```
1
5
10 1 10 1 10
```

<br>

### Sample Output

```
36
```

<br>

### Explanation

The maximum sum occurs when $A[1]=A[3]=A[5]=10$ and $A[2]=A[4]=1$. <br>
That is $|1 - 10| + |10 - 1| + |1 - 10| + |10 - 1| = 36$.<br>

# 풀이
<br>

S가 최대가 되는 지점은 B의 각 element별로 최대 혹은 최소(1)가 되는 지점들에서만 나올 수 있다.

<br>
<br>

````python
import os


def cost(B):
    dp = [[0], [0]]
    len_B = len(B)
    for i in range(1, len_B):
        dp[0].append(max(abs(B[i]-B[i-1])+dp[0][i-1], abs(B[i]-1)+dp[1][i-1]))
        dp[1].append(abs(1-B[i-1])+dp[0][i-1])
    return max(dp[0][-1], dp[1][-1])


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        n = int(input())

        B = list(map(int, input().rstrip().split()))

        result = cost(B)

        fptr.write(str(result) + '\n')

    fptr.close()
````
